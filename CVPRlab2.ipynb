### Understanding the fundamentals of transformation on images

```python
#Importing important libraries
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Reading the file
img = cv.imread('/content/Sample.jpg')
assert img is not None, "Image not found!"
```

```python
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)), plt.title("Original Image")
plt.show()
```

##Scaling the image using cv.resize

```python
scaled_img = cv.resize(img, None, fx=1.5, fy=1.5, interpolation=cv.INTER_LINEAR)
```

```python
plt.imshow(cv.cvtColor(scaled_img, cv.COLOR_BGR2RGB)), plt.title("Scaled Image")
plt.show()
```

##Rotating image by 45 degrees

```python
print(img.shape)
```

```python
(h, w) = img.shape[:2]
center = (w // 2, h // 2)
rotation_matrix = cv.getRotationMatrix2D(center, 45, 1)
rotated_img = cv.warpAffine(img, rotation_matrix, (w, h))
```

```python
plt.imshow(cv.cvtColor(rotated_img, cv.COLOR_BGR2RGB)), plt.title("Rotated Image")
plt.show()
```

Picture printed by Vemula Ananya (229310345)

##Shifting the image by [50,100]

```python
translation_matrix = np.float32([[1, 0, 50], [0, 1, 100]])
shifted_img = cv.warpAffine(img, translation_matrix, (w, h))
```

```python
plt.imshow(cv.cvtColor(shifted_img, cv.COLOR_BGR2RGB)), plt.title("Shifted Image")
plt.show()
```
