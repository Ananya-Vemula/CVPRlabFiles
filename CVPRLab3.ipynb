Log, Power Law, Negative and set operation transoformation on images



Importing Important libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('/content/funnybunny.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "Image not found or cannot be loaded."
```

```python
plt.imshow(img, cmap='gray'), plt.title("Original Image")
plt.show()
```

Log transformation

formula:
s = c*log(1+r)

```python
c_log = 255 / (np.log(1 + np.max(img)))
log_transformed = c_log * np.log(1 + img)
log_transformed = np.array(log_transformed, dtype=np.uint8)
```

```python
plt.imshow(log_transformed, cmap='gray'), plt.title("Log Transformation")
plt.show()
```

##Power law (Gamma Transformation)

Formula:
S=cR^

```python
gamma = 0.5
c_gamma = 255 / (np.power(255, gamma))
gamma_transformed = c_gamma * np.power(img, gamma)
gamma_transformed = np.array(gamma_transformed, dtype=np.uint8)
```

```python
plt.imshow(gamma_transformed, cmap='gray'), plt.title("Power-Law (Gamma) Transformation")
plt.show()
```

##Negative transformation

Formula:
n = 255 - r

```python
negative_transformed = 255 - img  # Inverts the pixel values
```

```python
plt.imshow(negative_transformed, cmap='gray'), plt.title("Negative Transformation")
plt.show()
```

##Applying set operations

```python
img1 = img
img2 = cv.imread('/content/WindowsOldBackground.jpg', cv.IMREAD_GRAYSCALE)
assert img1 is not None and img2 is not None, "One or both images could not be loaded."
```

```python
img2 = cv.resize(img2, (img1.shape[1], img1.shape[0]))
```

```python
plt.imshow(img1, cmap='gray'), plt.title("Image 1")
plt.show()
plt.imshow(img2, cmap='gray'), plt.title("Image 2")
plt.show()
```

```python
union = cv.bitwise_or(img1, img2)  # Union
```

```python
plt.imshow(union, cmap='gray'), plt.title("Union (OR)")
plt.show()
```

```python
intersection = cv.bitwise_and(img1, img2)  # Intersection
```

```python
plt.imshow(intersection, cmap='gray'), plt.title("Intersection (AND)")
plt.show()
```

```python
difference = cv.bitwise_xor(img1, img2)  # XOR (Symmetric Difference)
```

```python
plt.imshow(difference, cmap='gray'), plt.title("XOR (Difference)")
plt.show()
```

```python
subtraction = cv.bitwise_and(img1, cv.bitwise_not(img2))  # Difference (img1 - img2)
```

```python
plt.imshow(subtraction, cmap='gray'), plt.title("Difference (A - B)")
plt.show()
```
